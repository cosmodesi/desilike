# **Plotting examples**

In this notebook we will show how to make plots using the desilike plotting class, and also how to convert your chains onto getdist to make plots.

```python
#Read the chain via:
from desilike.samples import Chain
```

```python
#Import the chain from the 'fs_direct_examples.ipynb' file

chain = Chain.load('_tests/chain_fs_direct_0.npy').remove_burnin(0.3)
#Print the content of the chain object
print(chain)
```

# Desilike part

```python
#Import the plotting class from desilike
from desilike.samples import plotting
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 

#Lets make a plot for the 'qiso' param:
plotting.plot_triangle(chain, params=['h'], filled = True, markers=[0.6736])
```

```python
#We can use the cosmoprimo fiducial cosmology to put the markers:
from cosmoprimo.fiducial import DESI
cosmo = DESI() 

print(f'For example, the fiducial value of h: {cosmo["h"]}')
```

```python
#Lets see all the main cosmological parameters:
plotting.plot_triangle(chain, params=['h', 'omega_cdm', 'logA'], 
                       filled=True, markers=[cosmo['h'], cosmo['omega_cdm'],cosmo['logA']])
```

```python
#Now lets see the trace of the params:
plotting.plot_trace(chain);
```

```python
#We can convert the desilike chain to a getdist chain via:
getdist_chain = chain.to_getdist()
```

# Getdist part

```python
import getdist
from getdist import plots, MCSamples, loadMCSamples

#Initialize the plots objetc from getdist
g = plots.get_subplot_plotter()
#Print the triangle plot
g.triangle_plot(getdist_chain,filled=True,params=['h', 'omega_cdm', 'logA'],
                contour_colors=['tab:blue'],
                figure_kwargs={"figsize": (15,15)},
               markers = [cosmo['h'], cosmo['omega_cdm'],cosmo['logA']])
```
